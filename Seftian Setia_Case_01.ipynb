```python
#Import Libraries
import pandas as pd
import numpy as np
import scipy.stats as stats
```

```python
#Create DataFrame
#list / Crete Dictionary series
TransactionAmount = [100,150,50,100,130,120,100,85,70,150,150,120,50,100,100,140,90,150,50,90,120,100,110,75,65]
df = pd.DataFrame(TransactionAmount)
df
```

```python
#Central Tendency
import statistics
print ("Rata-rata /Mean data Transaction Amount")
print (statistics.mean(TransactionAmount))
print ("*******************************")
print ("Nilai Tengah /Median data Transaction Amount")
print (statistics.median(TransactionAmount))
```

```python
print (df.mode())
#cara2
print ("cara kedua :")
statistics.mode(TransactionAmount)
```

```python
#Tulis analisismu dari Central Tendency di sini
```

Dari data Central Tedency diatas maka bisa ditarik bahwa rata-rata Jumlah Transaksi penjualan adalah 100 maka para sales bisa mempunyai target 100 penjualan perbulan, lebih jauh lagi yang dibawah 100 penjulan bisa diberi sangsi dan yang di atas 100 bisa diberi bonus

```python
#Variability
#import ...
print ("Var : ",np.var(TransactionAmount))
print ("Std : ",np.std(TransactionAmount))
print ("Nilai Max: ", np.max(TransactionAmount), " & Nilai Min : ", np.min(TransactionAmount))
print ("Q1 : ",np.quantile(df, .25))
print ("Q2 : ",np.quantile(df, .50))
print ("Q3 : ",np.quantile(df, .75))
```

Distribusi datanya normal dari q1 ke q2 beda 15,q2 ke q3 beda 20 dari q3 ke q4 beda 30. bisa menjadi acuan performa terbaik yaitu nilai dengan capaian pejualan 120-150.

```python
#Define H0 and H1
#H0:
H0 ="HO rata-rata 102.6/100"
print(H0)
#H1:
H1 ="H1 Lebih besar dari rata-rata 102.6/100"
print(H1)
len(TransactionAmount)
```

```python
#Define alpha = 0.05
alpha = 0.05
```

```python
#Calculate t statistics and P-value
#Perform one sample t-test
stats.ttest_1samp(a=TransactionAmount, popmean=100)
```

```python
#statistic=0.4085001556802841 , pvalue=0.6865284813438117
```

```python
#Conclude with alpha and p-value
#Dengan P-value yang lebih besar dari Alpha 5% dan Pvalue 68% maka kita gagal menolak H0
#Conclusion: Transaction tidak efektif karena rata-rata masih di 100 transaksi
```

```python
#Conclude with t statistics and critical region
#Find the critical region
```

```python
#H1: lebih tinggi (masuk kanan)
#confidence interval
stats.t.ppf(1-0.05,24)
```
